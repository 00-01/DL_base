```python
from glob import glob
import os

import pandas as pd
```

```python
# data_dir = f"/media/z/0/MVPC10/DATA/v1.1/RAW/03"
number = "007"
data_dir = f"/home/z/{number}"
path = glob(f"{data_dir}/*/*.png")
path
```

```python
len(path)
```

```python
name = []
for i in path:
    name.append(int(os.path.basename(i).replace(".png", "")))

name
```

```python
df = pd.DataFrame(name, columns=['path'], dtype=int)
```
